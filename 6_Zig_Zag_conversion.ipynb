```python
input = "PAYPALISHIRING"
num_rows = 3
```

```python
def zigziag(input :str,num_rows :int):

    tensorlist = []
    i = 0

    for _ in range(len(input)):         
        while i  < num_rows:
            tensorlist.append([input[i:num_rows+i+2:2]])
            i += 1
        else:
            num_rows += i

    return tensorlist
```

```python
zigziag(input,num_rows)
```
